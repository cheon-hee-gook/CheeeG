    - 다국어 음성 인식 지원
    : 세계 최고 수준의 한국어/일본어 모델과 상용화 가능한 영어, 중국어, 프랑스어, 스페인어 모델을 제공

자율주행 분야 : __현대자동차__의 Smart ADAS 시스템
  - **ADAS 기술**
  : 크게 인지, 판단, 제어로 구분된다.
    - 인지는 센서나 카메라를 통해 환경을 인식.
    - 판단은 컨트롤러를 통해 신호를 처리하거나 주변 상황에 따라 차량의 움직임을 결정.
    - 가감속이나 조향제어 등 직접적인 움직임을 담당.

  - 차선 유지 지원 시스템 (LKAS, Lane Keeping Assist System)
  : 윈드실드 글라스에 장착된 카메라로 앞 차선을 인식한 후 방향 지시등 없이 차선을 이탈할 경우 스스로 경로를 바로잡음.

  - 자동 긴급 제동 시스템 (AEB, Autonomous Emergency Braking System)
  : 차간 간격과 주행 속도에 따른 충돌상황을 예측하고 브레이크를 작동시켜 추돌을 방지

  - 샤시 통합 제어 시스템 (VSM, Vehicle Stability Management System)
  : 제동시에 차량이 균형을 잃지 않도록 안전하게 자세를 유지.

  - 전방 추돌 경보 시스템 (FCWS, Forward Collision Warning System)
  : 전방감지카메라를 통해 차선과 선행차량을 인식하고, 추돌이 예상될때는 경보음과 스티어링 휠 진동으로 운전자에게 경고해 사고를 미연에 방지.

  - 어드밴스드 스마트 크루즈 컨트롤 (ASCC, Advanced Smart Cruise Control)
  : 상습 정체구간이나 혼잡한 도심에서 운전자가 페달과 브레이크를 번갈아 밟지 않아도 레이더 센서를 통해 앞 차량과의 거리를 일정하게 유지.
